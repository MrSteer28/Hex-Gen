**HEX GEN V1 - D3 SET - ONE INPUT, SINGLE LAYER, AUTOMATED SCALING.**

**HEX GEN V3 - EXPERIMENTAL - MULTIPLE INPUT OPTIONS**

```python
#Key variables.                                                                  !THESE NOTES WILL CAUSE FORMATTING ISSUES OUTSIDE OF COLAB!

print('Recommended grid size is between 1 and 10.')
grid_side = int(input('Select your grid size: '))                                #Input grid size and calculate required variables.
print('Recommended number of layers is 1 or 2.')
layers = int(input('Select number of layers: '))                                 #Input target number of layers.
layer_counter = 0                                                                #Counts completed layers.
grid_width = grid_side * 2 - 1                                                   #Target no. of columns (and maximum column height).
height_ref = grid_side                                                           #Variable target for column height.
column_counter = 0                                                               #Counts completed columns.
height_counter = 0                                                               #Counts completed hexes in column.
line = 150 / grid_side                                                           #The length of a segment (automated for scaling).
double = line * 2                                                                #Double segment for crossing hexes with one command.
print('Recommended pen size is between 2 and 15.')
print('(2 is optimal for larger grids and multiple layers.)')
pen_size = int(input('Select your pen size: '))                                  #Set pen size.
print('Hexagon sets: Default = Dist. Three Set, T = Three Set, S = Six Set,')
print('              G = Grid, A/B/C/D/E = Specific Hex Set, M = Manual.')
random_mode = input('Selected your hexagon set: ')                               #Select hex set and randomisation algorithm.
random_mode = random_mode.upper()
print('Colour modes: BLACK, RED, GREEN, BLUE, GREY,')
print('              RAINBOW, CUSTOM, LAYERS, GRADIENT.')
colour_mode = input('Select a colour mode: ')                                    #Select colour mode.
colour_mode = colour_mode.upper()

if random_mode == 'S':                                                           #Returns the number of three hexes drawn to give an indication of
  hex_count = (3 * grid_side * (grid_side - 1) + 1) * layers * 2                 #the processing time.
else:
    hex_count = (3 * grid_side * (grid_side - 1) + 1) * layers
print(f'\nHex count = {hex_count}')


#Initialise turtle and set parameters.

t.initializeTurtle(13) #Start turtle.
t.pu()
t.setposition(220,120) #Ideal position for these screen settings.
t.bgcolor('white')     #Set initial colour conditions.
t.pencolor('black')
t.rt(90)               #Set first heading.
t.pd()
t.hideturtle()         #Hide the turtle.
t.pensize(pen_size)

#Drawing

while layer_counter < layers:                                                    #Opens loop for multiple layers.

  if colour_mode == 'LAYERS':
    t.pencolor(r.choice(custom))

  if colour_mode == 'GRADIENT':
    t.pencolor((grey)[layer_counter])

  while column_counter < grid_width or height_counter < grid_side:               #Starts drawing loop for layer.

    if colour_mode != 'BLACK' and colour_mode != 'LAYERS' and \
       colour_mode != 'GRADIENT' and colour_mode != 'B':                         #Colours per hex according to colour mode.
      t.pencolor(r.choice(col_dic[colour_mode]))                                 #Selects random colour from library via dictionary.

    if random_mode == 'T':                                                       #Draws hex based on hex mode function.
      three_set()                                                                #Defaults to three_set_distributed.
    elif random_mode == 'S':
      layer_mode = 'N'
      six_set()
    elif random_mode == 'G':
      fullhex()
    elif random_mode == 'A':
      A_set()
    elif random_mode == 'B':
      B_set()
    elif random_mode == 'C':
      C_set()
    elif random_mode == 'D':
      D_set()
    elif random_mode == 'E':
      E1()
    elif random_mode == 'M':
      manual_set()

    else:
      three_set_distributed()

    height_counter += 1                                                          #Count hex in column height.

    if height_counter < height_ref:                                              #If column height not met,
      hexdrop()                                                                  #drop down and restart.

    elif column_counter + 1 < grid_side:                                         #If column height met before centre column,
      ascending()                                                                #add one to column target size and position to start.
      height_counter = 0                                                         #Counter must be + 1 as previous completion cannot be recorded
      column_counter += 1                                                        #until after check (interferes with previous loop).
      height_ref += 1

    else:                                                                        #If column height met on central column or after,
      descending()                                                               #reduce target column size and reposition correctly.
      column_counter += 1
      height_ref -= 1
      if column_counter < grid_width:                                            #Prevents infinite loop on height size by checking
        height_counter = 0                                                       #if column target has been reached.

  layer_counter += 1                                                             #Records layer completion.
  height_counter = 0                                                             #Resets counters and height_ref for next layer.
  column_counter = 0
  height_ref = grid_side
  t.pu()                                                                         #Sends turtle back to starting position.
  t.setposition(220,120)
  t.pd()
```
